```python
import numpy as np
import gym
```

```python
def initialization(env):
    values=np.zeros(env.nS)
    policy=np.zeros(env.nS)
    return values,policy
```

```python
def value_iteration(env,values,policy,threshold,discount):
    while True:
        delta=0
        for s in range(env.nS):
            value_list=[]
            for a in range(env.nA):
                act_value=0
                for transition, next_state, reward, done in env.P[s][a]:
                    act_value += transition * (reward + discount * values[next_state])
                value_list.append(act_value)
            value=np.max(value_list)
            policy[s]=np.argmax(value_list)
            delta=max(delta,np.abs(value-values[s]))
            values[s]=value
        if delta < threshold:
            break
    policy=policy.astype(np.int8)
    values=np.reshape(values, (4,4))
    print(values)
    return policy
```

```python
env=gym.make("FrozenLake-v0")
state=env.reset()
values,policy=initialization(env)
policy=value_iteration(env,values,policy,threshold=0.001,discount=0.9)
print(policy)
done=False
while not done:
    state, reward, done,_ = env.step(policy[state])
    env.render()
```
